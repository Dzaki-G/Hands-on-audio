```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
numpy_img_3ch = np.array(
[
    [[0, 120, 0], [0, 255, 0], [255, 0, 0]],
    [[255, 255, 0], [0, 255, 255], [255, 0, 255]],
    [[192, 192, 192], [128, 128, 128], [64, 64, 64]],
],
dtype=np.uint8
)
```

```python
plt.imshow(numpy_img_3ch)
plt.axis('off')  # sembunyikan sumbu
plt.show()
```

```python
# average all channels using numpy
gray_img_np = np.mean(numpy_img_3ch, axis=1).astype(np.uint8)
print(f"Isi Dari gray_img_np:\n{gray_img_np}")

plt.imshow(gray_img_np, cmap='gray')
plt.title('Grayscale Image (NumPy Average)')
plt.axis('off')  # Hide axis
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread(img_path)
plt.figure(figsize=(12, 6))
print(f"img.shape: {img.shape}")  # (height, width, channel)

# Cari koordinat tengah gambar
h, w, c = img.shape
center_y = h // 2
center_x = w // 2

# ===== TITIK MERAH DI TENGAH =====
img[center_y, center_x] = [0, 0, 255]  # BGR -> merah

# ===== KOTAK MERAH DI TENGAH =====
# Misalnya kotak 20x20 piksel
kotak_size = 10
img[center_y - kotak_size:center_y + kotak_size,
    center_x - kotak_size:center_x + kotak_size] = [0, 0, 255]

# Tampilkan
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Gambar dengan Titik/Kotak Merah di Tengah')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread(img_path)
img_conv = img.copy()

print(f"img.shape: {img.shape}")  # (height, width, channel)

# Titik tengah
titik_y_tengah = img.shape[0] // 2  # tinggi
titik_x_tengah = img.shape[1] // 2  # lebar

# Warnai kotak merah di tengah (BGR -> merah [0,0,255])
img_conv[titik_y_tengah-10:titik_y_tengah+10,
         titik_x_tengah-10:titik_x_tengah+10] = [0, 0, 255]

# Tampilkan
plt.figure(figsize=(8,8))
plt.imshow(cv2.cvtColor(img_conv, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Kotak Merah di Tengah (NumPy)')
plt.show()
```
